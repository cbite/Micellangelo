```python
import os
```

```python
root_dir = 'Input'
print('cleaning Input folder')
for root, dirs, files in os.walk(root_dir):

    for name in files:
        if not name.endswith('.ipynb'):
            print(os.path.join(root, name)) 
            os.remove(os.path.join(root, name))
print('\n\ndone')
```

```python
root_dir = 'Output'
print('cleaning Output folder')
for root, dirs, files in os.walk(root_dir):

    for name in files:
        if not name.endswith('.ipynb'):
            print(os.path.join(root, name)) 
            os.remove(os.path.join(root, name))
            
print('\n\ndone')
```
